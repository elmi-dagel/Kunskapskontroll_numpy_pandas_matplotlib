### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
- Figure: Hela fönstret där alla diagram ritas. Det kan innehålla flera diagram.
- Axes: Själva området där datan ritas upp (ett enskilt diagram). En figure kan ha flera axes.
- Axis: Själva x- eller y-axeln som visar skalan och gränserna för datan.
- Artist: Allt som ritas i ett diagram (som linjer, text eller etiketter) är en artist; det är byggstenen för allt i Matplotlib.
```

### When plotting in Matplotlib, what is the expected input data type?

```python
När man ritar diagram i Matplotlib förväntas följande datatyper som indata:
-Listor eller tuples: Vanliga Python-listor eller tuples med värden.
-NumPy-arrayer: Används ofta för effektiv hantering av numerisk data.
-Pandas DataFrames eller Series: Vanligt inom dataanalys för att plotta data.

Dessa datatyper används för att skapa olika typer av diagram som linjediagram och stapeldiagram.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())


# Random scatter data
np.random.seed(15)
x_scatter = np.random.randn(1000)
y_scatter = np.random.randn(1000)
x_line = np.linspace(-2, 2, 100)
y_line = x_line**2

# Bar data
fruit_counts = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
fruit_names = list(fruit_counts.keys())
fruit_values = list(fruit_counts.values())

# Create figure with subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Scatter plot
ax1.scatter(x_scatter, y_scatter)
ax1.set_title('Scatter Plot')
ax1.set_xlabel('Random X values')
ax1.set_ylabel('Random Y values')

# Bar plot with spaces
bar_width = 0.5  # Make the bars narrower
bar_positions = np.arange(len(fruit_names)) * 1.5  # Add space between bars
ax2.bar(bar_positions, fruit_values, width=bar_width)
ax2.set_title('Fruit Count')
ax2.set_xlabel('Fruit')
ax2.set_ylabel('Count')
ax2.set_xticks(bar_positions)
ax2.set_xticklabels(fruit_names)

# Display the plots
plt.show()





```
